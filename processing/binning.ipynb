```python
import os
import json

import pandas as pd
import numpy as np
```

```python
os.chdir("..")
```

# Parameters

```python
antibiotic_dictionary_file = open("data/antibiotics.json", "r")
antibiotic_dictionary = json.load(antibiotic_dictionary_file)
labels = antibiotic_dictionary["Escherichia coli"]

bin_size = 40
```

```python
file = "data/unprocessed/raw/e_coli_driams.csv"

bac = pd.read_csv(file)
bac
```

# Binning

```python
malditof = bac[bac.columns.drop(list(bac.filter(regex='[^0-9]')))]
malditof
```

```python
binned_malditof = malditof.T.groupby(np.arange(len(malditof.columns)) // bin_size).mean().T
binned_malditof.columns = [col for col in range(int(malditof.columns[0]), int(malditof.columns[-1]), bin_size)]
binned_malditof
```

```python
binned_bac = pd.DataFrame(np.column_stack([binned_malditof, bac[labels]]))
binned_bac.columns = list(binned_malditof.columns) + labels
binned_bac
```

# Saving File

```python
file_name_ext = os.path.basename(file)
file_name = os.path.splitext(file_name_ext)[0]
ext = os.path.splitext(file_name_ext)[1]
new_file_name = "data/unprocessed/binned/"+file_name+"_bin"+str(bin_size)+ext

binned_bac.to_csv(new_file_name, index=False, mode='w')
```
